```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.style.use('ggplot')

%matplotlib inline
```

```python
#前処理
```

```python
kg = pd.read_csv("kaggle_housing_price.csv")
```

```python
# データを最初の5行だけ表示
kg.head(5)
```

```python
# データの行数、列数を表示
kg.shape
```

```python
# 要約統計量を表示
kg.describe()
```

```python
# インデックス番号0~10の行を抽出
kg.ix[0:10]
```

```python
# "SalePrice"と"LotArea"の列を抽出し、最初の10行だけ表示
kg.ix[:10][["SalePrice","LotArea"]]
```

```python
# "1stFlrSF"と"2ndFlrSF"を合計した"FlrSF_total"を新たな列として加える
kg["FlrSF_total"] = kg["1stFlrSF"] + kg["2ndFlrSF"]
```

```python
# 列'SaleType'をダミー変数に展開したものを変数 dataset に上書きします。
dataset = pd.get_dummies(kg,columns=["SaleType"])
```

```python
# ダミー変数が作成されていることを確認します　(datasetの最初の5行だけ出力)
dataset.head(5)
```

```python
# 'YearBuilt'が2000以降の物件のみを抽出し、最初の5件のみ表示
kg.query("YearBuilt>=2000")
```

```python
# 'YearBuilt'が2000以降、'GarageCars'が2以上の物件を抽出
kg.query("YearBuilt>=2000 and GarageCars>=2")
```

```python
# datasetの'SalePrice'をヒストグラムで表示
dataset["SalePrice"].hist()
```

```python
# datasetの'LotArea'と'SalePrice'を散布図で表示
dataset.plot(kind="scatter",x="LotArea",y="SalePrice")
```

```python
#'SalePrice'のSaleCondition毎の平均を変数　price_by_conditionに格納
price_by_condition=kg.groupby("SaleCondition").aggregate({"SalePrice" : np.mean}).reset_index()

# price_by_conditionが持つ、棒グラフを表示する命令を実行
price_by_condition.plot.bar(x = "SaleCondition")
```

```python
# datasetの'SaleCondition'ごとに'SalePrice'をboxplotで表示
dataset[["SaleCondition","SalePrice"]].boxplot(by="SaleCondition")
```

```python
# 列ごとに欠損値の有無を確認
aa= kg.columns
for bb in aa :
    cc = sum(pd.isnull(kg[bb]))
print(aa+":"+ str(cc))
```
